```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
from statsmodels.formula.api import ols
import matplotlib.ticker as ticker
from matplotlib import cm
import squarify  # for treemap

# Set the style for better aesthetics
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_context("notebook", font_scale=1.2)

# Load your dataset (replace with actual path)
df = pd.read_excel("Fraud data FY 2023-24 for B&CC.xlsx")

# Data preprocessing
# Convert date columns to datetime
df['POLICYRISKCOMMENCEMENTDATE'] = pd.to_datetime(df['POLICYRISKCOMMENCEMENTDATE'], errors='coerce')
df['Date of Death'] = pd.to_datetime(df['Date of Death'], errors='coerce')
df['INTIMATIONDATE'] = pd.to_datetime(df['INTIMATIONDATE'], errors='coerce')

# Create date difference columns
df['Policy_to_Death_Days'] = (df['Date of Death'] - df['POLICYRISKCOMMENCEMENTDATE']).dt.days
df['Death_to_Intimation_Days'] = (df['INTIMATIONDATE'] - df['Date of Death']).dt.days

# Add month and year columns for time series analysis
df['Month_Year'] = df['INTIMATIONDATE'].dt.to_period('M').astype(str)
df['Year'] = df['INTIMATIONDATE'].dt.year
```

```python
# 1. State-wise Fraud Count - Horizontal Bar Chart
plt.figure(figsize=(12, 8))
state_counts = df['CORRESPONDENCESTATE'].value_counts().nlargest(10)
ax = sns.barplot(x=state_counts.values, y=state_counts.index, palette='viridis')
plt.title('Top 10 States with Highest Fraud Cases', fontsize=16, fontweight='bold')
plt.xlabel('Number of Fraud Cases', fontsize=12)
plt.ylabel('State', fontsize=12)

# Add count values to the bars
for i, v in enumerate(state_counts.values):
    ax.text(v + 0.5, i, f"{v}", va='center', fontsize=10)

plt.tight_layout()
plt.savefig('state_fraud_count.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 2. Channel Analysis - Bar Chart with Custom Colors
plt.figure(figsize=(12, 6))
channel_counts = df['CHANNEL'].value_counts()
sns.barplot(x=channel_counts.index, y=channel_counts.values, palette='plasma')
plt.title('Fraud Count by Channel', fontsize=16, fontweight='bold')
plt.xlabel('Channel', fontsize=12)
plt.ylabel('Number of Fraud Cases', fontsize=12)
plt.xticks(rotation=45, ha='right')

# Add count values on top of bars
for i, v in enumerate(channel_counts.values):
    plt.text(i, v + 0.5, f"{v}", ha='center', fontsize=10)

plt.tight_layout()
plt.savefig('channel_fraud_count.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 3. Time Series Analysis - Line Chart
plt.figure(figsize=(14, 6))
time_series = df.groupby(df['INTIMATIONDATE'].dt.to_period('M').astype(str)).size()
time_series = time_series.reset_index()
time_series.columns = ['Month', 'Count']
time_series['Month'] = pd.to_datetime(time_series['Month'])
time_series = time_series.sort_values('Month')

plt.plot(time_series['Month'], time_series['Count'], marker='o', linestyle='-',
         linewidth=2, markersize=8, color='navy')
plt.title('Monthly Trend of Fraud Cases', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Fraud Cases', fontsize=12)
plt.grid(True, alpha=0.3)
plt.xticks(rotation=45)

# Format x-axis to show months nicely
plt.gca().xaxis.set_major_formatter(ticker.DateFormatter('%b %Y'))

plt.tight_layout()
plt.savefig('time_series_fraud.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 4. Distribution Analysis - Histograms
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Policy to Death histogram
sns.histplot(df['Policy_to_Death_Days'].dropna(), bins=30, kde=True, color='steelblue', ax=ax1)
ax1.set_title('Days Between Policy Start and Death', fontsize=14, fontweight='bold')
ax1.set_xlabel('Days', fontsize=12)
ax1.set_ylabel('Frequency', fontsize=12)
median_val = df['Policy_to_Death_Days'].median()
ax1.axvline(median_val, color='red', linestyle='--', linewidth=1.5)
ax1.text(median_val*1.1, ax1.get_ylim()[1]*0.9, f'Median: {median_val:.0f} days',
         color='red', fontsize=10)

# Death to Intimation histogram
sns.histplot(df['Death_to_Intimation_Days'].dropna(), bins=30, kde=True, color='indianred', ax=ax2)
ax2.set_title('Days Between Death and Intimation', fontsize=14, fontweight='bold')
ax2.set_xlabel('Days', fontsize=12)
ax2.set_ylabel('Frequency', fontsize=12)
median_val = df['Death_to_Intimation_Days'].median()
ax2.axvline(median_val, color='red', linestyle='--', linewidth=1.5)
ax2.text(median_val*1.1, ax2.get_ylim()[1]*0.9, f'Median: {median_val:.0f} days',
         color='red', fontsize=10)

plt.tight_layout()
plt.savefig('distribution_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 5. Top Fraud Hotspots - Horizontal Bar Chart
fraud_data = df[df['Fraud Category'].notna()]
fraud_counts = fraud_data.groupby(['CORRESPONDENCEPOSTCODE', 'CORRESPONDENCECITY']).size().reset_index(name='Fraud_Count')
top_frauds = fraud_counts.sort_values(by='Fraud_Count', ascending=False).head(10)
top_frauds['Location'] = top_frauds['CORRESPONDENCECITY'] + ' (' + top_frauds['CORRESPONDENCEPOSTCODE'].astype(str) + ')'

plt.figure(figsize=(12, 8))
colors = sns.color_palette('viridis', len(top_frauds))
bars = plt.barh(top_frauds['Location'], top_frauds['Fraud_Count'], color=colors)
plt.title('Top 10 Fraud Hotspots', fontsize=16, fontweight='bold')
plt.xlabel('Number of Fraud Cases', fontsize=12)
plt.ylabel('Location (City & Postcode)', fontsize=12)

# Add count values to the bars
for i, (bar, value) in enumerate(zip(bars, top_frauds['Fraud_Count'])):
    plt.text(value + 0.3, i, f"{value}", va='center', fontsize=10)

plt.tight_layout()
plt.savefig('top_fraud_hotspots.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 6. Relationship Analysis - Scatter Plot with Regression Lines by Channel
plt.figure(figsize=(12, 8))

# Filter out NaN values
scatter_df = df.dropna(subset=['Policy_to_Death_Days', 'Death_to_Intimation_Days', 'CHANNEL'])

# Create scatter plot with regression lines for each channel
channels = scatter_df['CHANNEL'].unique()
colors = sns.color_palette('tab10', len(channels))

for i, channel in enumerate(channels):
    channel_data = scatter_df[scatter_df['CHANNEL'] == channel]

    # Plot points
    plt.scatter(channel_data['Policy_to_Death_Days'],
                channel_data['Death_to_Intimation_Days'],
                alpha=0.6, label=channel, color=colors[i], s=40)

    # Add regression line if there are enough points
    if len(channel_data) > 2:
        # Fit regression line
        x = channel_data['Policy_to_Death_Days']
        y = channel_data['Death_to_Intimation_Days']

        # Get line of best fit
        m, b = np.polyfit(x, y, 1)

        # Create line using min and max x values
        x_line = np.array([min(x), max(x)])
        y_line = m * x_line + b

        # Plot regression line
        plt.plot(x_line, y_line, '-', color=colors[i], linewidth=2)

plt.title('Relationship Between Policy-to-Death and Death-to-Intimation Intervals',
          fontsize=16, fontweight='bold')
plt.xlabel('Days Between Policy Start and Death', fontsize=12)
plt.ylabel('Days Between Death and Intimation', fontsize=12)
plt.legend(title='Channel')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('relationship_analysis.png', dpi=300, bbox_inches='tight')
plt.show()  # Added display command
```

```python
# 7. Channel Distribution by State - Stacked Bar Chart
state_channel = df.groupby(['CORRESPONDENCESTATE', 'CHANNEL']).size().reset_index(name='Count')
pivot_data = state_channel.pivot(index='CORRESPONDENCESTATE', columns='CHANNEL', values='Count').fillna(0)
top_states = pivot_data.sum(axis=1).sort_values(ascending=False).head(10).index

# Filter for top 10 states
pivot_top10 = pivot_data.loc[top_states]

plt.figure(figsize=(14, 8))
pivot_top10.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Channel Distribution in Top 10 States', fontsize=16, fontweight='bold')
plt.xlabel('State', fontsize=12)
plt.ylabel('Number of Fraud Cases', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Channel', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('channel_distribution_by_state.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 8. Treemap for Geographical Distribution
plt.figure(figsize=(14, 10))

# Prepare data for treemap
state_counts = df.groupby('CORRESPONDENCESTATE').size().reset_index(name='Count')
state_counts = state_counts.sort_values('Count', ascending=False)

# Normalize for color mapping
state_counts['NormCount'] = (state_counts['Count'] - state_counts['Count'].min()) / (state_counts['Count'].max() - state_counts['Count'].min())

# Generate colors
cmap = cm.get_cmap('viridis')
state_counts['Color'] = state_counts['NormCount'].apply(lambda x: cmap(x))

# Create treemap
squarify.plot(sizes=state_counts['Count'],
              label=state_counts['CORRESPONDENCESTATE'] + '\n' + state_counts['Count'].astype(str),
              color=state_counts['Color'],
              alpha=0.8,
              pad=True)

plt.axis('off')
plt.title('Geographical Distribution of Fraud Cases', fontsize=16, fontweight='bold')

plt.tight_layout()
plt.savefig('geographical_treemap.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Calculate metrics by channel
radar_data = df.groupby('CHANNEL').agg(
    Avg_Policy_to_Death=('Policy_to_Death_Days', 'mean'),
    Avg_Death_to_Intimation=('Death_to_Intimation_Days', 'mean'),
    Count=('CHANNEL', 'size')
).reset_index()

# Normalize the metrics for radar chart
for col in ['Avg_Policy_to_Death', 'Avg_Death_to_Intimation', 'Count']:
    max_val = radar_data[col].max()
    min_val = radar_data[col].min()
    if max_val != min_val:  # Avoid division by zero
        radar_data[f'{col}_Normalized'] = (radar_data[col] - min_val) / (max_val - min_val)
    else:
        radar_data[f'{col}_Normalized'] = 1.0

# Get only the top 5 channels by count for clarity
top_channels = radar_data.nlargest(5, 'Count')

# Set up radar chart
categories = ['Avg Days\nPolicy to Death', 'Avg Days\nDeath to Intimation', 'Number of Cases']
n_categories = len(categories)

# Create angles for radar chart
angles = np.linspace(0, 2*np.pi, n_categories, endpoint=False).tolist()
angles += angles[:1]  # Close the circle

fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))

# Plot each channel
for i, row in top_channels.iterrows():
    values = [row['Avg_Policy_to_Death_Normalized'],
              row['Avg_Death_to_Intimation_Normalized'],
              row['Count_Normalized']]
    values += values[:1]  # Close the polygon

    ax.plot(angles, values, linewidth=2, label=row['CHANNEL'])
    ax.fill(angles, values, alpha=0.25)

# Customize radar chart
ax.set_theta_offset(np.pi / 2)  # Start from top
ax.set_theta_direction(-1)  # Go clockwise
ax.set_thetagrids(np.degrees(angles[:-1]), categories)
ax.set_ylim(0, 1)
ax.set_yticks([])  # Remove radial ticks
ax.grid(True)
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
plt.title('Multi-dimensional Comparison of Top 5 Channels', fontsize=16, fontweight='bold')

plt.tight_layout()
plt.savefig('radar_chart.png', dpi=300, bbox_inches='tight')
plt.show()  # Added display command

```
